```python
import pandas as pd
```

```python
df_csv = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dataset_doencas_cardiacas/heart - heart.csv')
print(df_csv)
```

```python
# Verificar tipos de variáveis
df_csv.dtypes
```

```python
# checagem de valores nulos na base de dados

df_csv.isnull().sum()
```

```python
# Calcular a média da Idade
media_idade = df_csv['Age'].mean()
print(f'Média de Idade: {media_idade}')
```

```python
# Calcular a média de colesterol por tipo de dor no peito
media_colesterol = df_csv.groupby('ChestPainType')['Cholesterol'].mean()
print('Média de Colesterol por Tipo de Dor no Peito:')
print(media_colesterol)
```

```python
# Contar a quantidade de casos de doença cardíaca por sexo
contagem_doenca = df_csv.groupby('Sex')['HeartDisease'].sum()
print('Número de Casos de Doença Cardíaca por Sexo:')
print(contagem_doenca)
```

```python
# Selecionar apenas colunas numéricas
colunas_numericas = df_csv.select_dtypes(include=['float64', 'int64'])

# Calcular a matriz de correlação
correlacao = colunas_numericas.corr()
print('Matriz de Correlação:')
print(correlacao)
```

```python
# Calculando média, mediana e moda
media_cholesterol = df_csv['Cholesterol'].mean()
mediana_cholesterol = df_csv['Cholesterol'].median()
moda_cholesterol = df_csv['Cholesterol'].mode()

# Resultados
print(f"Média do Colesterol: {media_cholesterol}")
print(f"Mediana do Colesterol: {mediana_cholesterol}")
print(f"Moda do Colesterol: {moda_cholesterol.tolist()}")
```

```python
# desvio padrão da coluna idade
desvio_padrao_age = df_csv['Age'].std()
print(f"Desvio Padrão Idade: {desvio_padrao_age}")
```

```python
# resumo estatístico das colunas numéricas
print(df_csv.describe())
```
